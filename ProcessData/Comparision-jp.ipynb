Jupyter Notebook Comparison. Processing of IMU data by using rotation matrix

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

not_locked = ['time', 'pelvis_rotation', 'lumbar_extension', 'lumbar_bending',
       'lumbar_rotation', 'arm_flex_r', 'arm_add_r', 'arm_rot_r',
       'elbow_flex_r', 'arm_flex_l', 'arm_add_l', 'arm_rot_l', 'elbow_flex_l']
```

```python
rpi_filter = pd.read_csv(r"C:\Users\chris\Documents\Visualisation\Opensim\Results\23-04-03_1757\ik_tiny_file.mot",sep='\t',skiprows=6)
jp_filter = pd.read_csv(r"C:\Users\chris\Documents\Visualisation\Opensim\Results\23-04-03_1757\ik_processed_tiny_file.mot",sep='\t',skiprows=6)
```

```python
ranges = []
for col in not_locked:
    ranges.append((col,jp_filter[col].max() - jp_filter[col].min()))

ranges.sort(key = lambda x: x[1],reverse = True)
for x in ranges[:5]:
    print(x)
```

Compare the first 4 items

```python
x = jp_filter["time"]
y_variable = 'elbow_flex_l'

plt.figure(figsize=(20,6))

y = rpi_filter[y_variable]
plt.plot(x,y,label = 'RPI')

y = jp_filter[y_variable]
plt.plot(x,y,label = 'com')

Title = "Variation of " + y_variable +" filtering on RPI vs Jupyter"

plt.title(Title)


plt.legend()
plt.show()
```
